#### Preliminary manipulations:

1. Add the age according to the formula $Age_{yr} = \#Rings + 1.5$
2. Vectorize the `'Sex'` variable:
    + Infant (I) will be `0`
    + Male (M) will be `1`
    + Female (F) will be `2`

```python
# Add a column with the age
aba['Age'] = aba['Rings'] + 1.5

# Define the transformation to apply to the sexes
def vectSex(x):
    if x == 'F':
        return 2
    elif x == 'M':
        return 1
    else:
        return 0

# Apply said transformation, putting the result in a new column
aba['VSex'] = aba['Sex'].apply(lambda x: vectSex(x))
```

### 2 - Data Exploration (Plotting 1/3)

Exploration of the relations between `'Length'`, `'Diameter'` and `'Height'` against `'Age'`:

```python
# Define plotting function for a pair plot
def pairPlt(alpha, mode = 'Normal'):
    g = sns.PairGrid(aba, y_vars = ['Age'], x_vars = ['Length', 'Diameter', 'Height'], hue = 'Sex')
    g.map(plt.scatter, alpha = alpha)
    if mode == 'log':
        for ax in g.axes.flat:
            ax.set_xscale('log', basex = 2)
            ax.set_yscale('log', basey = 2)
    g.add_legend()

# Define plotting function for a facet grid
def facetPlt(var, mode = 'Normal'):
    g = sns.FacetGrid(aba, col = 'Sex')
    g.map(plt.scatter, var, 'Age', alpha = 0.5)
    if mode == 'log':
        for ax in g.axes.flat:
            ax.set_xscale('log', basex = 2)
            ax.set_yscale('log', basey = 2)
```

### 3 - Linear Regression

First, create a dedicated dataset for experimentation:

```python
experiment = pd.concat([aba['Sex'],
                        aba['VSex'],
                        aba['Diameter']], axis = 1)
experiment['DiameterLog'] = np.log2(experiment['Diameter'])
experiment['DiameterLogSqu'] = experiment['DiameterLog'] ** 2
experiment['DiameterLogCub'] = experiment['DiameterLog'] ** 3
experiment['Length'] = aba['Length']
experiment['LengthLog'] = np.log2(experiment['Length'])
experiment['LengthLogSqu'] = experiment['LengthLog'] ** 2
experiment['LengthLogCub'] = experiment['LengthLog'] ** 3
experiment['WeightWhole'] = aba['Weight_Whole']
experiment['WeightWholeLog'] = np.log2(aba['Weight_Whole'])
experiment['WeightShucked'] = aba['Weight_Shucked']
experiment['WeightShuckedLog'] = np.log2(aba['Weight_Shucked'])
experiment['Age'] = aba['Age']
experiment['AgeLog'] = np.log2(experiment['Age'])

pd.set_option('display.max_columns', 11)
```

### 3 - Linear Regression

Split the experimental dataset into:

* 80% of the observations $\Rightarrow$ training data
* 20% of the observations $\Rightarrow$ test data

```python
SPLIT = 0.8

# Get the indexes for the training data and the test data
train_idx = sample(experiment.index, int(SPLIT * len(experiment)))
test_idx = experiment.index.difference(train_idx)
```

And create some features set:

```python
ft00 = ['VSex', 'Diameter']
ft01 = ['VSex', 'DiameterLog']
ft02 = ['VSex', 'DiameterLog', 'DiameterLogSqu']
ft03 = ['VSex', 'DiameterLog', 'DiameterLogSqu', 'DiameterLogCub']
ft04 = ['VSex', 'Length']
ft05 = ['VSex', 'LengthLog']
ft06 = ['VSex', 'LengthLog', 'LengthLogSqu']
ft07 = ['VSex', 'LengthLog', 'LengthLogSqu', 'LengthLogCub']
ft08 = ['VSex', 'Diameter', 'Length', 'WeightWhole']
ft09 = ['VSex', 'Diameter', 'Length', 'WeightShucked']
ft10 = ['VSex', 'DiameterLog', 'LengthLog', 'WeightWholeLog']
ft11 = ['VSex', 'DiameterLog', 'LengthLog', 'WeightShuckedLog']
ft12 = ['VSex', 'Diameter', 'Length', 'WeightWholeLog']
ft13 = ['VSex', 'Diameter', 'Length', 'WeightShuckedLog']
tg = 'AgeLog'

ft_list = [ft00, ft01, ft02, ft03, ft04, ft05, ft06, ft07, ft08, ft09, ft10, ft11, ft12, ft13]
```

Try to fit a linear model with all the features' set and compare the results:

```python
def linearModelTest(ft_list, index_training, index_testing):
    model = linear_model.LinearRegression()
    
    train_results = []
    test_results = []

    for idx, fts in enumerate(ft_list):
        X_train = experiment.ix[index_training, fts].values
        y_train = experiment.ix[index_training, tg]
        X_test = experiment.ix[index_testing, fts].values
        y_test = experiment.ix[index_testing, tg]
        model.fit(X_train, y_train)
        train_results.append(model.score(X_train,y_train))
        test_results.append(model.score(X_test,y_test))
        print('Features ft{:0>2d}: score (train) = {:.4f} score (test) = {:.4f} SSE = {}'.
              format(idx, model.score(X_train, y_train), model.score(X_test, y_test), np.sum((model.predict(X_train)-y_train) ** 2)))
    return (train_results, test_results)
```

```python
train_series_lin, test_series_lin = linearModelTest(ft_list, train_idx, test_idx)
```

### 4 - Polynomial Regression

```python
def polyModelTest(ft_list, index_training, index_testing):
    best_score = 0.0
    best_train = []
    best_test = []
    alphas = [1,2,5,10,20,50,100,200,500,1000]
    
    for idx, fts in enumerate(ft_list):
        X_train = experiment.ix[index_training, fts].values
        y_train = experiment.ix[index_training, tg]
        X_test = experiment.ix[index_testing, fts].values
        y_test = experiment.ix[index_testing, tg]
        
        for alpha in alphas:
            model = linear_model.Ridge(alpha = alpha)
            model.fit(X_train, y_train)
            if model.score(X_train,y_train) > best_score:
                best_score = model.score(X_train,y_train)
                best_train.append(best_score)
                best_test.append(model.score(X_test, y_test))
                print('Features ft{:0>2d} & alpha = {:<4d}: score (train) = {:.4f} score (test) = {:.4f} SSE = {}'.
                      format(idx, alpha, model.score(X_train, y_train), model.score(X_test, y_test), 
                             np.sum((model.predict(X_train)-y_train) ** 2)))
    return (best_train, best_test)
```

```python
train_series, test_series = polyModelTest(ft_list, train_idx, test_idx)
train_series = pd.Series(train_series)
test_series = pd.Series(test_series)
```

```python
train_plot = train_series.plot()
test_plot = test_series.plot()
plt.legend(('Training', 'Testing'));
plt.xticks(range(len(train_series)));
```
